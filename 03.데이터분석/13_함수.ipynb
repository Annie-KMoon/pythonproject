```python
def add_cm(height):
    if isinstance(height, int):
        return f'{height} cm'
    else:
        return int(height.replace('cm',''))
```

```python
print(add_cm('165cm'))
```

```python
def lower(sw):
    if pd.notnull(sw):
        return sw.lower()
    else:
        return "없음"
```

```python
df_score=df.loc[:,'국어':'사회']
df_score['총점']=df_score.apply(lambda row:sum(row['국어':'사회']), axis=1)
df_score['평균']=df_score.apply(lambda row:row['총점']/5, axis=1)
df_score.loc['과목평균']=df_score.apply(np.mean, axis=0) #행추가
df_score
```
