```python
import pandas as pd
df = pd.read_csv('data/score.csv')
df.head() #5개 데이터만
df.set_index('지원번호', inplace=True)
df[:3]
```

```python
df.groupby('학교').get_group('능남고')
df.groupby('학교').get_group('북산고')
```

```python
df2 = df.groupby('학교').size().to_frame('학생수')
df2
```

```python
df3 = df.groupby('학교')['키'].mean().to_frame('평균키')
```

```python
df3 = df.groupby('학교')[['국어','영어','수학']].mean()
df3.columns=['국어평균', '영어평균','수학평균']
df3
```

```python
df2 = df.groupby('학교').count() #Null값외 카운트
df2[['이름','SW특기']]
```

```python
df.loc['4번','SW특기']='java'
df.loc['5번','SW특기']='python'
df['SW특기']=df['SW특기'].str.capitalize()
df['SW특기'].unique()
```

```python
df.groupby('SW특기').size().to_frame()
```

```python
df2 = df['SW특기'].value_counts(normalize=True).to_frame()
df2['퍼센트'] = df2['proportion']*100
df2
```
