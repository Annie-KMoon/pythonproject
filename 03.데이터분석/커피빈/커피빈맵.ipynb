```python
import requests
import numpy as np

def get_location(address):
    url = "https://dapi.kakao.com/v2/local/search/address.json?query="+address
    headers = {'Authorization':'KakaoAK 3b1f91160f090c6816a03b93720659b4'}
    res = requests.get(url, headers=headers).json()
    if res['documents']:
        x=res['documents'][0]['x']
        y=res['documents'][0]['y']

    else:
        x = np.nan
        y = np.nan
    return x,y
    
```

```python
x,y = get_location('서울시 강남구 영동대로 607')
x,y
```

```python
store = pd.read_csv('커피빈매장.csv')
store['X']=''
store['Y']=''
store.head(2)
```

```python
for i in store.index:
    address = store.loc[i,'Address']
    x,y = get_location(address)
    store.loc[i, 'X']=x
    store.loc[i, 'Y']=y
```

```python
#주소부분일치 지도 출력
store.to_csv('커피빈맵.csv')
address = '마포구'
filt = store['Address'].str.contains(address)
df = store[filt]
df.head()

import folium
import warnings
warnings.filterwarnings(action='ignore')
def get_map(address):
    filt = store['Address'].str.contains(address)
    df = store[filt]
    if len(df.index) ==0:
        print("찾을 수 없습니다.")
    else:
        first = df.head(1)
        x=first['X']
        y=first['Y']
        location = (y,x)
        map = folium.Map(location, zoom_start=10, width='100%', height='100%')

        for idx, row in df.iterrows():
            x = row['X']
            y = row['Y']
            name=row['Name']

            folium.Marker(location=[y,x],popup=folium.Popup(name, max_width=200),
                          icon=folium.Icon(color='red', icon='star')).add_to(map)
        map.save(f'{address}.html')

get_map('상암')
```
