```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
#데이터 읽기 라이브러리 import
import pandas as pd
enroll = pd.read_csv('data/수강.csv')
enroll
```

```python
group = enroll.groupby('강좌번호').size().to_frame('수강인원')
cou = pd.read_csv('data/강좌.csv')
merge = cou.merge(group, on='강좌번호')
merge
```

```python
labels = merge['강좌이름']
values = merge['수강인원']
labels, values
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels,explode=[0.02]*len(values),
        autopct='%.1f%%', colors=colors,
        wedgeprops= {'width':0.7, 'edgecolor':'brown'})
plt.show()
```

```python
#성적별 등급 함수
def grade(score):
    if score>=90:
        return "A"
    elif score>=80:
        return "B"
    elif score>=70:
        return "C"
    elif score>=60:
        return "D"
    else:
        return "F"
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll['학점'] = enroll['점수'].apply(grade) #점수하나컬럼으로 처리-axis 불필요
group = enroll.groupby('학점').size()
group
```

```python
labels = group.index
values = group.values
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels+'학점',explode=[0.02]*len(group),
        autopct='%.0f%%', colors=colors,
        wedgeprops= {'width':0.65, 'edgecolor':'ivory'},
        textprops={'color':'gray'})
plt.show()
```
