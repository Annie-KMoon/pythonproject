```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
#데이터 읽기 라이브러리 import
import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
#원그래프-label, value / %비율확인시 사용
#SW특기자수 저체 인원 중 몇명인지

group = df.groupby('SW특기').size()
group
```

```python
labels = group.index
values = group.values

labels, values
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90,  counterclock=True)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 
```

```python
plt.pie(values, labels=labels,
        autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=[0.03]*len(values),
        autopct='%.1f%%', startangle=90, counterclock=False)
plt.legend(loc=(1.1,0.3))
plt.show()
```

```python
#2%미만측정치 생략 함수
def custom_autopct(pct):
    if pct <=2:
        return ""
    else:
        return f'{pct:.0f}%'
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, explode=[0.03]*len(values),colors=colors,
        wedgeprops={'width':0.6, 'edgecolor':'gray', 'linewidth':2},
        autopct=custom_autopct, startangle=90, counterclock=False)
plt.legend(loc=(1.1,0.3))
plt.show()
```

```python
#특기자별 원그래프
group = df.groupby('SW특기').size()
labels = group.index
values = group.values
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, autopct='%.1f%%',colors=colors)
plt.show()
```
