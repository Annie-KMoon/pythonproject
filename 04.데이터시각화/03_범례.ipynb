```python
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Noto Sans KR')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
year = [year for year in range(2020, 2025, 1)] #연속값-반복문사용
height = [145, 150, 160, 165, 172]
```

```python
plt.plot(year, height, label='년도별 신장')
plt.legend(loc = 'upper left') #범례
plt.legend(loc=(1.1,0.5)) #숫자 위치로도 지정 가능
plt.show()
```
