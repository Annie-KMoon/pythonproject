```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Noto Sans KR')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
name = df['이름']
height = df['키']
```

```python
plt.barh(name, height) 
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160,210)
plt.show()
```

```python
plt.figure(figsize=(10,5))
bar = plt.bar(name,height, color='none', ec='brown')
plt.ylim(160,210)
bar[0].set_hatch('/')
bar[1].set_hatch('..')
bar[2].set_hatch('|')
bar[3].set_hatch('-')
bar[4].set_hatch('o')
bar[5].set_hatch('--')
bar[6].set_hatch('-.')
bar[7].set_hatch('*')
for idx, h in enumerate(height):
    plt.text(idx,h+1, h, ha='center')
plt.show()
```

```python
#학생별평균
name = df['이름']
kor = df['국어']
eng = df['영어']
mat = df['수학']
df['평균'] = df.apply(lambda row:row['국어':'사회'].sum()/5, axis=1)
avg=df['평균']

plt.figure(figsize=(10,5))
plt.title('학생별평균')
plt.ylim(0,100)
plt.xlabel('학생이름')
plt.ylabel('평균점수')
plt.yticks([y for y in range(0,100,5)])
bar = plt.bar(name,avg,label='평균')
for idx, a in enumerate(avg):
    plt.text(name[idx],avg[idx]+2,a, color='navy', ha='center', size=10)
plt.show()
```
