```python
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Noto Sans KR')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
x = [1,2,3,4,5]
y = [3,6,9,12,15]
```

```python
plt.plot(x,y)
plt.xticks([ x for x in range(min(x), max(x)+1, 1)], color = 'blue', size = 15)
plt.yticks([ y for y in range(min(y), max(y)+1, 2)])
plt.show()
```
