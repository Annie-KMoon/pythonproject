```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
#데이터 읽기 라이브러리 import
import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
#다중막대그래프 위치 산출
x = df.index
x = x*3
x
```

```python
#다중막대그래프
x = df.index*4
plt.bar(x+0, df['국어'], label='국어', width=1)
plt.bar(x+1, df['영어'], label='영어', width=1)
plt.bar(x+2, df['수학'], label='수학', width=1)
xticks=[idx+1 for idx in x] #중앙위치
plt.xticks(xticks, df['이름']) #이름으로변경

for idx in df.index:
    x = idx*4
    kor=df.loc[idx,'국어']
    plt.text(x, kor+1, f'{kor}점', color='gray', ha='center', size=8)

for idx in df.index:
    x = idx*4
    eng=df.loc[idx,'영어']
    plt.text(x+1, eng+1, f'{eng}점', color='gray', ha='center', size=8)

for idx in df.index:
    x = idx*4
    mat=df.loc[idx,'수학']
    plt.text(x+2, mat+1, f'{mat}점', color='gray', ha='center', size=8)
    
            
plt.legend(loc=(1.1,0.2))
plt.show()
```

```python
#다중막대그래프 사회과학
plt.figure(figsize=(10,4))
x = df.index*3
plt.bar(x+0, df['사회'], label='사회', width=1, color='#5F9EA0')
plt.bar(x+1, df['과학'], label='과학', width=1, color='#FFE4E1')
xticks=[idx+0.5 for idx in x] #중앙위치
plt.xticks(xticks, df['이름']+'\n'+df['지원번호']) #이름으로변경

for idx in df.index:
    x = idx*3
    so=df.loc[idx,'사회']
    plt.text(x, so+1, f'{so}점', color='gray', ha='center', size=8)
    sci=df.loc[idx,'과학']
    plt.text(x+1, sci+1, f'{sci}점', color='gray', ha='center', size=8)

    
plt.legend(loc=(1.02,0))
plt.show()
```
