```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Noto Sans KR')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
name = df['이름']
height = df['키']
```

```python
#막대그래프 그리기
colors = ['r','orange','yellow','green','b','navy','violet','purple']
plt.bar(name, height, color=colors, alpha=0.5) #alpha:투명도
plt.show()
```

```python
plt.bar(name, height, width=0.5)
plt.ylim(160,210)
plt.yticks([y for y in range(160,211, 5)])
plt.xticks(name, df.index, rotation=90)
plt.show()
```
