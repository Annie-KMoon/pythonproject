```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
#데이터 읽기 라이브러리 import
import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
#학생별(이름) 국어/영어/수학점수 꺽은선 그래프
import pandas as pd
df = pd.read_csv('data/score.csv')
name=df['이름']
height=df['키']
plt.figure(figsize=(10,3))
plt.title('학생별 국어, 영어, 수학 점수',size=15)
plt.ylim(0,110)
plt.plot(df['이름'], df['국어'], label='국어',
         marker ='o', ms=7, mec='gray', mew=3, mfc='yellow')
plt.plot(df['이름'], df['영어'], label='영어',lw=3, ls='--', color='r')
plt.plot(df['이름'], df['수학'], label='수학', marker ='o', ms=7, mec='gray', mew=3, mfc='yellow')
yticks = [y for y in range(0,101,50)]
# plt.yticks(yticks, [f'{y}점' for y in yticks]) #???
xticks=[f'{df.loc[x,"이름"]}\n{df.loc[x,"지원번호"]}'for x in range(len(df))]
plt.xticks(df['이름'], xticks)
plt.xlabel('이름', size=15, loc='right', color='g')
plt.ylabel('점수', size=15, loc='top', color ='g')
plt.legend(loc=(0.5,-0.3), ncol=3)
plt.show()
```

```python
#학생별(이름) 키 막대그래프
import pandas as pd
df = pd.read_csv('data/score.csv')
name=df['이름']
height=df['키']
plt.figure(figsize=(10,3))
plt.bar(name, height, color='#B0C4DE', hatch='o', ec='#F0EBA3')
plt.ylim(160,210)
yticks=[y for y in range(160,211,10)]
plt.yticks=(yticks, [f'{y}cm' for y in yticks])
xticks=[f'{df.loc[x,"이름"]}\n{df.loc[x,"지원번호"]}'for x in range(len(df))]
plt.xticks(df['이름'], xticks)
plt.show()
```

```python
#학생별 평균점수 막대그래프
df['평균']=df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
plt.bar(df['이름'], df['평균'])
avg=df['평균']
for idx, a in enumerate(avg):
    plt.text(idx, a+1, f'{a:.2f}점', color='navy', ha='center', size=10)
plt.show()
```
