```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 8)
plt.rc('axes', unicode_minus = False)
```

```python
#행정구역별 인구수
df = pd.read_excel('행정구역_202401.xlsx', skiprows=1)
df.head(3)
# df.info()
```

```python
df1 = df.copy()
df1.columns = ['행정구역2','총인구수','남자인구수','여자인구수']
df1['행정구역1'] = '' #행추가
df1.head(2)
```

```python
#데이터 행 정리
df2 = df1.copy()
cols = list(df1.columns)
df2 = df1[[cols[-1]] + cols[0:-1]] #=df1[['행정구역1','행정구역2'...]] 행옮기기
df2.head()
```

```python
#행 채워서 정리
sidos = ['특별시','광역시','자치시','경기도','자치도','북도','남도']
for i in range(1,len(df2)):
    for sido in sidos:
        if sido in df2.loc[i, '행정구역2']:
            df2.loc[i,'행정구역1'] = df2.loc[i, '행정구역2']
            df2.loc[i,'행정구역2'] = '소계'
df2.head()
```

```python
#행정구역1 빈칸 채우기
for i in range(2,len(df2)):
    if df2.loc[i,'행정구역1']=='':
        df2.loc[i,'행정구역1'] = df2.loc[i-1,'행정구역1']
df2.loc[0,'행정구역1']='전국'
df2.loc[0,'행정구역2']='합계'
df2['행정구역2'] = df2['행정구역2'].str.strip()
df2
```

```python
df2.to_csv('행정구역별인구수.csv') #파일저장
```

```python
group = df3.groupby('행정구역1')['총인구수'].sum()
df4 = group.to_frame('총인구수')
sort = df4.sort_values('총인구수', ascending= False) #인구 내림차순
df5 = sort[:11]
df5 = df5.sort_index()
df5
```

```python
plt.figure(figsize=(20,8))
plt.bar(df5.index, df5['총인구수'])
plt.xticks(df5.index, rotation = 45)
plt.show()
```
