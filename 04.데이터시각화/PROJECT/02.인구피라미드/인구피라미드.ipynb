```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#전국 연령별 인구현황(구글-주민조사-다운로드)-남자데이터만 처리
#불필요한값외 필요값만 정리
df_m = pd.read_excel('201508.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m = df_m.head(1)
#문자-숫자로변경
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_m.iloc[0,0]
df_m
```

```python
#전국 연령별 인구현황(구글-주민조사-다운로드)-여자데이터만 처리
#불필요한값외 필요값만 정리
df_w = pd.read_excel('201508.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w = df_w.head(1)
#문자-숫자로변경
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w.iloc[0,0]
df_w
```

```python
#컬럼명 통일
df_w.columns = df_m.columns
df_w.columns
```

```python
plt.figure(figsize=(10,7))
plt.title('2015년 대한민국 성별 인구 피라미드\n')
plt.barh(df_m.columns, -df_m.iloc[0]//1000)
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.savefig('2015년.png')
plt.show()
```
