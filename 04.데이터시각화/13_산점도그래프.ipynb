```python
#그래프생성 라이브러리 import
import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
#데이터 읽기 라이브러리 import
import pandas as pd
df = pd.read_csv('data/score.csv')
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['국어'],df['영어'], sizes=df['학년']*100,
             c = df['학년'], cmap='GnBu', alpha=0.3)
plt.xlabel('국어')
plt.ylabel('영어')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)
plt.show()
```

```python
plt.scatter(df['국어'],df['영어'], sizes=df['학년']*400,
             c = df['학년'], cmap='GnBu', alpha=0.3)
plt.xlabel('국어')
plt.ylabel('영어')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)

for idx in range(len(df)):
    x=df.loc[idx,'국어']
    y=df.loc[idx,'영어']
    name=df.loc[idx,'이름']
    plt.text(x,y,name, size=8, ha='center')
plt.show()
```

```python
#수학:과학
plt.scatter(df['수학'],df['과학'], sizes=df['학년']*500,
             c = df['학년'], cmap='coolwarm_r', alpha=0.2)
plt.xlabel('수학')
plt.ylabel('과학')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)

for idx in range(len(df)):
    x=df.loc[idx,'수학']
    y=df.loc[idx,'과학']
    name=df.loc[idx,'이름']
    plt.text(x,y,name, size=8, ha='center')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv')
df['학년'] = [3,3,2,1,1,3,2,2]
plt.scatter(df['국어'],df['영어'],sizes=df['학년']*500,
             c = df['학년'], cmap='autumn_r', alpha=0.3)
plt.xlabel('국어')
plt.ylabel('영어')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)

for idx in range(len(df)):
    x=df.loc[idx,'국어']
    y=df.loc[idx,'영어']
    name=df.loc[idx,'이름']
    plt.text(x,y,name, size=6, ha='center')

plt.ylim(0,120)
    
plt.show()
```
