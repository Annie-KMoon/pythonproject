```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt
#한글폰트 설정
plt.rc('font', family = 'Noto Sans KR')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
name = list(df['이름']) #시리즈로 사용 가능
height = list(df['키'])
name, height
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름', loc = 'right', fontdict={'size':13})
plt.ylabel('키(cm)', loc = 'top', fontdict={'size':13})
# plt.xticks(name)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth=2, marker='o')
plt.show()
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름', loc = 'right', fontdict={'size':13})
plt.ylabel('키(cm)', loc = 'top', fontdict={'size':13})
# plt.xticks(name)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth=2, marker='o', markersize =10, 
         markerfacecolor='yellow', markeredgecolor='gray', markeredgewidth = 2)
plt.show()
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름', loc = 'right', fontdict={'size':13})
plt.ylabel('키(cm)', loc = 'top', fontdict={'size':13})
# plt.xticks(name)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, 
         linewidth=2, linestyle= ':', color = 'gray',
         marker='o', markersize =10, 
         markerfacecolor='yellow', markeredgecolor='gray', markeredgewidth = 2)
plt.show()
```

```python
#스타일 축약어
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름', loc = 'right', fontdict={'size':13})
plt.ylabel('키(cm)', loc = 'top', fontdict={'size':13})
# plt.xticks(name)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height,
         lw=3, c='#708090', ls=':', 
         marker='o', ms=10, mfc='#6A5ACD', mec='#2F4F4F', mew=3)
plt.show()
```

```python
#그래프 사이즈 지정
plt.figure(figsize=(10,3))
plt.ylim(160, 210)
plt.plot(name, height, marker='o')
for idx, txt in enumerate(height):
    plt.text(name[idx],height[idx]+2, f'{txt}cm', size=8, ha='center')
plt.show()
```
